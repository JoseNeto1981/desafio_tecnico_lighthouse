## Questao 7 

O presente estudo tem como objetivo desenvolver um modelo preditivo simples para estimar a demanda diária de um produto específico, com base em dados históricos de vendas.

O problema apresentado envolve a necessidade de melhorar o planejamento de estoque, evitando tanto a falta de produtos (resultando em perda de vendas) quanto o excesso de estoque (gerando custos desnecessários).

Para isso, foi utilizado um modelo baseline baseado em média móvel de 7 dias, respeitando as seguintes premissas:

- O período de treino considera dados até 31/12/2023
- O período de teste corresponde a todo o mês de janeiro de 2024
- A previsão é realizada em base diária
- Não há utilização de dados futuros (evitando data leakage)
- A análise é restrita ao produto: **Motor de Popa Yamaha Evo Dash 155HP**

Além disso, como a base de vendas não contém o nome do produto, foi necessário utilizar uma tabela auxiliar de produtos para identificar o `ID_PRODUCT` correspondente.

##  Importação de Bibliotecas e Conexão com o Snowflake e leitura das bases de dados


Nesta etapa, são importadas as bibliotecas necessárias e estabelecida a conexão com o banco de dados Snowflake para acesso à tabela de vendas.
Os dados de vendas são carregados a partir do Snowflake, enquanto a tabela de produtos é lida a partir de um arquivo CSV auxiliar. Esta tabela é necessária para identificar o código do produto alvo.

```python
import pandas as pd
import os

from dotenv import load_dotenv
from snowflake.snowpark import Session
from sklearn.metrics import mean_absolute_error


load_dotenv()

conn = {
    "account": os.getenv("account"),
    "user": os.getenv("user"),
    "password": os.getenv("password"),
    "role": "ACCOUNTADMIN",
    "warehouse": "COMPUTE_WH",
    "database": "LH_NAUTICALS"
}

session = Session.builder.configs(conn).create()


# 2. LER TABELA DE VENDAS

df_snowpark = session.table("LH_NAUTICALS.QUESTAO7.VENDAS_PREVISAO")
df_vendas = df_snowpark.to_pandas()

# padronizar nomes das colunas
df_vendas.columns = df_vendas.columns.str.strip().str.upper()

print("Colunas de vendas:")
print(df_vendas.columns)

print("\nHead vendas:")
print(df_vendas.head())

# 3. LER PLANILHA DE PRODUTOS

df_snowpark = session.table("LH_NAUTICALS.QUESTAO7.PRODUTOS_RAW")
df_produtos = df_snowpark.to_pandas()


# padronizar nomes das colunas
df_produtos.columns = df_produtos.columns.str.strip().str.upper()

print("\nColunas de produtos:")
print(df_produtos.columns)

print("\nHead produtos:")
print(df_produtos.head())
```

 Ajuste de Tipos de Dados

Nesta etapa, são realizados ajustes nos tipos de dados das colunas, garantindo que:

- As colunas numéricas estejam no formato adequado para cálculos
- A coluna de data seja corretamente interpretada como datetime
- Os códigos dos produtos estejam no mesmo tipo para permitir o relacionamento entre as tabelas

Esse tratamento é essencial para evitar inconsistências e erros ao longo da modelagem.



```python
## Ajustar tipos
df_vendas["ID_PRODUCT"] = pd.to_numeric(df_vendas["ID_PRODUCT"], errors="coerce")
df_vendas["QTD"] = pd.to_numeric(df_vendas["QTD"], errors="coerce")
df_vendas["TOTAL"] = pd.to_numeric(df_vendas["TOTAL"], errors="coerce")
df_vendas["SALE_DATE"] = pd.to_datetime(df_vendas["SALE_DATE"], dayfirst=True, errors="coerce")

df_produtos["CODE"] = pd.to_numeric(df_produtos["CODE"], errors="coerce")
```

##  Identificação do Produto Alvo

Como a tabela de vendas não contém o nome do produto, foi necessário buscar o `ID_PRODUCT` correspondente ao produto solicitado utilizando a tabela de produtos.

```python
## Encontrar o produto alvo

produto_alvo = "Motor de Popa Yamaha Evo Dash 155HP"

produto_encontrado = df_produtos[
    df_produtos["NAME"].str.contains(produto_alvo, case=False, na=False)
].copy()

print("\nProduto encontrado:")
print(produto_encontrado)

# validação
if produto_encontrado.empty:
    raise ValueError("Produto alvo não encontrado na planilha de produtos.")

if len(produto_encontrado) > 1:
    print("\nAtenção: mais de um produto encontrado. Verifique manualmente:")
    print(produto_encontrado[["CODE", "NAME"]])

# pegar o código do produto
id_produto_alvo = produto_encontrado.iloc[0]["CODE"]

print(f"\nID_PRODUCT encontrado para o produto alvo: {id_produto_alvo}")
```

##  Filtragem e Tratamento dos Dados

Nesta etapa, os dados são filtrados para o produto alvo e convertidos para os tipos adequados, garantindo consistência na análise.

```python
## Filtrar vendas dos produtos

df_produto = df_vendas[df_vendas["ID_PRODUCT"] == id_produto_alvo].copy()

print("\nQuantidade de linhas de vendas do produto alvo:")
print(len(df_produto))

print("\nAmostra das vendas do produto:")
print(df_produto.head())

# remover linhas inválidas
df_produto = df_produto.dropna(subset=["SALE_DATE", "QTD"])

print("\nMenor data:", df_produto["SALE_DATE"].min())
print("Maior data:", df_produto["SALE_DATE"].max())
```

## Construção da Série Temporal Diária

Os dados são agregados por dia e os dias sem venda são preenchidos com valor zero, garantindo continuidade temporal para o modelo.

```python
## Agregando as vendas por dia
df_diario = (
    df_produto.groupby("SALE_DATE", as_index=False)["QTD"]
    .sum()
    .sort_values("SALE_DATE")
)

# transformar em série diária completa
df_diario = df_diario.set_index("SALE_DATE").asfreq("D", fill_value=0)
df_diario = df_diario.rename(columns={"QTD": "VENDAS"})

print("\nSérie diária:")
print(df_diario.head(10))
print(df_diario.tail(10))
```

##  Separação entre Treino e Teste

A base foi dividida em:
- Treino: até 31/12/2023
- Teste: janeiro de 2024

Essa separação garante que não haja uso de dados futuros.

```python
## Treino e teste
treino = df_diario.loc[: "2023-12-31"].copy()
teste = df_diario.loc["2024-01-01":"2024-01-31"].copy()

print("\nÚltima data treino:", treino.index.max())
print("Primeira data teste:", teste.index.min())
print("Última data teste:", teste.index.max())
print("Qtd dias teste:", len(teste))

# validação
if treino.empty:
    raise ValueError("O conjunto de treino ficou vazio. Verifique as datas do produto.")
if teste.empty:
    raise ValueError("O conjunto de teste ficou vazio. Verifique se existem dados para janeiro de 2024.")
```

##  Modelo Baseline — Média Móvel de 7 Dias

Foi implementado um modelo simples baseado na média dos últimos 7 dias, utilizando apenas dados passados para evitar data leakage.

```python
## Previsão média móvel 7 dias
previsoes = []
historico = treino["VENDAS"].copy()

for data in teste.index:
    previsao_dia = historico.tail(7).mean()
    previsoes.append(previsao_dia)

    # adiciona o valor real ao histórico somente após prever o dia
    valor_real = teste.loc[data, "VENDAS"]
    historico.loc[data] = valor_real

teste["PREVISAO_BASELINE"] = previsoes
```

##  Avaliação do Modelo e Resultados das Previsões

A qualidade das previsões foi avaliada utilizando a métrica MAE (Mean Absolute Error).
A tabela abaixo apresenta os valores reais e previstos para cada dia de janeiro de 2024.

```python
## Calcular o MAE
mae = mean_absolute_error(teste["VENDAS"], teste["PREVISAO_BASELINE"])

print(f"\nMAE do baseline: {mae:.2f}")

## RESULTADO FINAL

resultado = teste.reset_index().rename(columns={
    "SALE_DATE": "DATA",
    "VENDAS": "REAL",
    "PREVISAO_BASELINE": "PREVISTO"
})

print("\nPrevisões diárias de janeiro de 2024:")
print(resultado)
```

```python
print("\nResposta objetiva:")
print("a. O baseline é adequado?")
print("   O baseline pode servir como referência inicial, mas sua adequação depende da estabilidade das vendas do produto.")
print("   Se houver muita oscilação diária, o modelo tende a apresentar erro mais alto.")

print("\nb. Limitação do método:")
print("   A média móvel de 7 dias não captura tendência, sazonalidade, promoções ou eventos externos.")
```

# Previsão primeira semana de janeiro

Foi calculada a soma das previsões entre os dias 01/01 e 07/01 de 2024

```python
primeira_semana = teste.loc["2024-01-01":"2024-01-07"]

soma_previsao = primeira_semana["PREVISAO_BASELINE"].sum()

soma_previsao_arredondada = round(soma_previsao)

print("\nSoma da previsão (01/01 a 07/01):", soma_previsao_arredondada)

print(primeira_semana[["VENDAS", "PREVISAO_BASELINE"]])
```

##  Visualização das Previsões

A seguir, é apresentado um gráfico comparando os valores reais e previstos ao longo do período de teste, permitind

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(teste.index, teste["VENDAS"], label="Real")
plt.plot(teste.index, teste["PREVISAO_BASELINE"], label="Previsto")

plt.title("Real vs Previsto - Janeiro 2024")
plt.xlabel("Data")
plt.ylabel("Vendas")
plt.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

##  Conclusão

O modelo baseline apresentou um erro médio absoluto (MAE) baixo, indicando boa aderência aos dados, principalmente devido à predominância de dias sem vendas.

Entretanto, observou-se que o modelo não consegue antecipar picos de demanda, como ocorreu em alguns dias específicos, pois se baseia apenas na média dos últimos 7 dias.

### Respostas objetivas:

**a. O baseline é adequado?**  
O modelo pode ser considerado adequado como ponto de partida, pois captura bem períodos sem vendas. No entanto, sua capacidade de prever picos é limitada.

**b. Limitação do método:**  
A média móvel não captura sazonalidade, tendências ou eventos pontuais, resultando em baixa capacidade de prever variações abruptas na demanda.
