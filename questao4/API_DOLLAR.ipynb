# API Dólar (Questão 4)
Notebook criado para capturar evolução do dólar usando a API do Banco Central.

```python
import requests
import pandas as pd

url = "https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/CotacaoDolarPeriodo(dataInicial=@dataInicial,dataFinalCotacao=@dataFinalCotacao)?@dataInicial='01-01-2023'&@dataFinalCotacao='12-31-2024'&$top=100000&$format=json"

response = requests.get(url, timeout=60)
response.raise_for_status()

data = response.json()
df = pd.DataFrame(data["value"])

df["data_cambio"] = pd.to_datetime(df["dataHoraCotacao"]).dt.date

df_final = (
    df.groupby("data_cambio", as_index=False)["cotacaoVenda"]
      .mean()
      .rename(columns={"cotacaoVenda": "cotacao_venda"})
)

df_final.to_csv("cambio_bcb.csv", index=False, encoding="utf-8")
print(df_final.head())
print("Arquivo salvo: cambio_bcb.csv")
```
